```python
class Node:
    def __init__(self, data: list, level, f_val):
        self.data = data
        self.level = level
        self.f_val = f_val

    def generate_children(self):
        children = []
        values = [[0, 1], [1, 0], [-1, 0], [0, -1]]
        row, col = self.find()
        for offset in values:
            child = self.move_blank(row, col, row + offset[0], col + offset[1])
            if child:
                child_node = Node(child, self.level + 1, 0)
                children.append(child_node)
        return children

    def move_blank(self, row, column, row_new, column_new):
        if 0 <= row_new < len(self.data) and 0 <= column_new < len(self.data):
            temp = self.copy()
            t = temp[row][column]
            temp[row][column] = temp[row_new][column_new]
            temp[row_new][column_new] = t
            return temp
        else:
            return None

    def find(self):
        for row in range(len(self.data)):
            for column in range(len(self.data)):
                if self.data[row][column] == '_':
                    return row, column

    def copy(self):
        res = []
        for row in self.data:
            temp = []
            for elem in row:
                temp.append(elem)
            res.append(temp)
        return res


def calc_h_value(node, goal):
    temp = 0
    for i in range(len(node.data)):
        for j in range(len(node.data)):
            if node.data[i][j] != goal[i][j]:
                temp += 1
    return temp


def calc_f_value(node, goal):
    return node.level + calc_h_value(node, goal)


def solve(start, goal):
    open_list = []
    closed_list = []
    st = Node(start, 0, 0)
    st.f_val = calc_f_value(st, goal)
    open_list.append(st)
    while True:
        curr: Node = open_list[0]
        if calc_h_value(curr, goal) == 0:
            closed_list.append(curr)
            break
        for i in curr.generate_children():
            if i not in closed_list and open_list:
                i.f_val = calc_f_value(i, goal)
                open_list.append(i)
        closed_list.append(curr)
        del open_list[0]
        open_list.sort(key=lambda x: x.f_val)
    return closed_list


goal_outer = [[1, 2, 3], [4, 5, 6], [7, 8, '_']]
ans = solve([[1, 2, 3], [4, '_', 5], [7, 8, 6]], goal_outer)
for ind in ans:
    print(ind.data)
```
